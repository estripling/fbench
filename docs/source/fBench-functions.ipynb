# Overview of fBench functions

## Version check

```python
import bumbag
import matplotlib.pyplot as plt
import numpy as np

import fbench

print(fbench.__version__)
```

```python
num = 101
settings_scatter = dict(
    s=60,
    c="black",
    marker="X",
    alpha=0.6,
    zorder=2,
)

for name, config in fbench.FUNCTION_CONFIGURATIONS.items():
    print(bumbag.headline(name))

    x_coord = np.linspace(min(config.x_bounds), max(config.x_bounds), num)
    y_coord = np.linspace(min(config.y_bounds), max(config.y_bounds), num)
    coord = fbench.create_coordinates3d(config.func, x_coord, y_coord)

    fig = plt.figure(figsize=plt.figaspect(0.4))
    ax = fig.add_subplot(1, 2, 1, projection="3d")
    ax = fbench.create_surface_plot(coord, ax=ax)
    ax.scatter(*config.global_minimum_x, config.global_minimum_fx, **settings_scatter)

    ax = fig.add_subplot(1, 2, 2)
    ax = fbench.create_contour_plot(coord, ax=ax)
    ax.scatter(*config.global_minimum_x, **settings_scatter)
    ax.axis("scaled")

    plt.tight_layout()
    plt.show()
```
