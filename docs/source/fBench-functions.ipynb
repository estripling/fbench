## Ackley 

```python
coord = fbench.create_coordinates3d(fbench.ackley, np.linspace(-5, 5, 101))
```

```python
ax = fbench.create_contour_plot(coord)
ax.axis("scaled")
plt.tight_layout()
plt.show()
```

```python
fig, ax = fbench.create_surface_plot(coord)
plt.tight_layout()
plt.show()
```
